```python
import requests
import pandas as pd
```

```python
url = "https://content.guardianapis.com/technology/elon-musk?from-date=2024-01-01&api-key=5d3d7dec-7e16-41a5-b2e3-9a19f03b6e82&type=article&page=1"


response = requests.get(url)
```

```python
if response.status_code == 200:
    # Extract data in JSON format
    data = response.json()

    articles = data['response']['results']

    # Create lists to store data
    headlines = []
    publication_dates = []
    urls = []


    for article in articles:
        headlines.append(article['webTitle'])
        publication_dates.append(article['webPublicationDate'])
        urls.append(article['webUrl'])


    df = pd.DataFrame({
        'Headline': headlines,
        'Publication Date': publication_dates,
        'URL': urls
    })


    print(df.head())

else:
    print("Failed to fetch data. Status code:", response.status_code)
```

```python
df.to_csv('Extract elon musk news.csv', index=False)
```
